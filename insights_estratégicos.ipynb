```python
# üìä An√°lise Explorat√≥ria de Produtos - Dataset Tratado

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
caminho_arquivo = 'Dados_tratados.csv'
df = pd.read_csv("data/Dados_tratados.csv")

# -----------------------------------------------
# üîç Etapa 1: Correla√ß√£o entre Vari√°veis Num√©ricas
# -----------------------------------------------
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correla√ß√£o entre Vari√°veis Num√©ricas")
plt.show()

# ---------------------------------------------------------
# üîé Etapa 2: Produtos com Alta Nota e Poucas Avalia√ß√µes
# ---------------------------------------------------------
produtos_ocultos = df[(df['Nota'] >= 4.5) & (df['N_Avalia√ß√µes'] < 10)]

# Visualizar os principais produtos com esse perfil
print("Produtos com Nota >= 4.5 e N_Avalia√ß√µes < 10:")
print(produtos_ocultos[['T√≠tulo', 'Nota', 'N_Avalia√ß√µes', 'Qtd_Vendidos', 'Pre√ßo']].sort_values(by='Nota', ascending=False))

# Gr√°fico de dispers√£o
plt.figure(figsize=(10, 6))
sns.scatterplot(data=produtos_ocultos, x='Pre√ßo', y='Nota', hue='Qtd_Vendidos', palette='viridis', size='N_Avalia√ß√µes', sizes=(40, 200))
plt.title('Produtos Bem Avaliados com Poucas Vendas')
plt.xlabel('Pre√ßo')
plt.ylabel('Nota')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
plt.show()

# ------------------------------------------------
# üìå A√ß√µes Estrat√©gicas Recomendadas (Resumo Final)
# ------------------------------------------------
acoes = """
üéØ A√ß√µes Estrat√©gicas Recomendadas:

1. ‚úÖ Refor√ßar o marketing de produtos bem avaliados com poucas avalia√ß√µes (potencial escondido).
2. üì¶ Aumentar o estoque dos produtos com muitas vendas e avalia√ß√µes positivas.
3. üì¢ Investir em campanhas promocionais nas categorias mais populares.
4. üïµÔ∏è Explorar padr√µes sazonais ou impacto dos descontos para ajustar estrat√©gias de pre√ßo.
5. üõçÔ∏è Verificar se produtos com pre√ßo alto possuem avalia√ß√µes ruins ‚Äì pode indicar necessidade de ajustes.
"""

print(acoes)
```
